```python
import matplotlib.pyplot as plt
import numpy as np

news_list, rec_list = [409, 251, 86], [3621314, 1447890, 607568]
bars, color = ['Left', 'Center', 'Right'], ['#08519c','#a8a495','#9a0200']
x_pos = np.arange(len(bars))

def remove_spines(ax):
    ax.spines['right'].set_visible(False)
    ax.spines['top'].set_visible(False)

# Add arrows to axes
def add_arrows(ax):
#     ax.annotate("", xy=(1.1, 0), xycoords='axes fraction', xytext=(0, 0),
#                 arrowprops=dict(arrowstyle="->", lw=1.5))
    ax.annotate("", xy=(0, 1.1), xycoords='axes fraction', xytext=(0, 0),
                arrowprops=dict(arrowstyle="->", lw=1.25))

plt.figure(figsize=(8,2),dpi=400)
ax1 = plt.subplot(121)
remove_spines(ax1)
add_arrows(ax1)
plt.bar(x_pos, news_list, color=color, width=0.5)
plt.xticks(x_pos, bars, fontproperties='Arial', size=10, weight='bold')
plt.xlim((-0.5,2.5))
plt.ylim((0,450))
plt.tick_params(direction='in')
plt.yticks(fontproperties='Arial', size=10, weight='bold')
# plt.hlines([500, 1000], xmin=-1,xmax=3, color='grey', ls='-', lw=0.2)
plt.ylabel('#Articles', fontproperties='Arial', size=10, weight='bold')

ax2 = plt.subplot(122)
remove_spines(ax2)
add_arrows(ax2)
plt.bar(x_pos, np.array(rec_list)/1000000, color=color, width=0.5)
plt.xticks(x_pos, bars, fontproperties='Arial', size=10, weight='bold')
plt.xlim((-0.5,2.5))
# plt.ylim((0,4))
plt.tick_params(direction='in')
plt.yticks(fontproperties='Arial', size=10, weight='bold')
plt.ylabel('#Recommendations', fontproperties='Arial', size=10, weight='bold')
# plt.savefig('./figures/manual_articles_num_1B1C.pdf', dpi=400, bbox_inches='tight', pad_inches=0)
plt.show()
```
