```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
from scipy import stats
import numpy as np

source_all = pd.read_csv('../../data/source_all_june.csv').set_index('i')
manual = source_all[['source','manual']]
manual = manual[manual['manual']<66]
manual.reset_index(inplace=True)
all_manual_ratings = manual['manual'].values

fig, ax = plt.subplots(figsize=(28,6), tight_layout=True)
n, bins, patches = ax.hist(all_manual_ratings, np.linspace(-0.7,0.7,50), color='green', histtype='bar', edgecolor='black')
ax.set_ylim(0,60)
ax.set_xlim(-0.75,0.75)
# ax.bar(x1_list, click_list, width=width, color='#5154A6', align='edge', label='#Click', edgecolor='black')
for c,p in zip(np.arange(len(n)),patches):
    if c<21:
        plt.setp(p,'facecolor','#08519c')
    elif c<28:
        plt.setp(p,'facecolor','#a8a495')
    else:
        plt.setp(p,'facecolor','#9a0200')

ax.set_ylabel('#News', fontsize=28, fontweight='bold')
ax.axvspan(-0.75,-0.1,facecolor='blue',alpha=0.1)
ax.axvspan(-0.1,0.1,facecolor='grey',alpha=0.1)
ax.axvspan(0.1,0.75,facecolor='red',alpha=0.1)
plt.yticks([0,20,40,60], fontproperties='Arial', weight='bold', size=25)
plt.xticks(np.arange(-0.7,0.71,0.1), fontproperties='Arial', weight='bold', size=25)
plt.savefig('./figures/manual_distribution_S3.pdf', dpi=400,bbox_inches='tight',pad_inches=0)

plt.show()
```
